The purpose of this notebook is to graph pH data for lactobacilli growing on different mucin glycans.

```python
from elliot_utils import *
from matplotlib.legend_handler import HandlerTuple
```

```python
figPath = Path.cwd().joinpath('figures/glycans_pH/')
analysisPath = Path.cwd().joinpath('analysis_files/glycans_pH/')
```

```python
data = pd.read_csv(analysisPath.joinpath('72hrs_pH_data.csv'))
```

```python
# Try graphing pH as a function of growth
fig, ax = plt.subplots(figsize=(4, 6))
plot = sns.scatterplot(data=data.drop(data[data.Strain != 'L. crispatus MV-1A-US'].index), x='Delta OD600', y='pH', hue='Glycan', s=150)
plot.set(title='$\it{L. crispatus}$ MV-1A-US ($\it{nanE}$+)\n', xlabel='Growth (ΔOD600)', ylabel='Spent Media pH', xlim=(0, 1.2), ylim=(3.9, 7.1))
plt.legend(title='Glycan', bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
sns.despine()
#saveFig(plot, figPath.joinpath('Lcrispatus_MV1AUS.png'))
```

```python
# Graph the data from the lactic acid enzyme assay
lactateData = pd.read_csv(analysisPath.joinpath('lactic_acid_data.csv'))
```

```python
plot = sns.catplot(data=lactateData, x='Species', y='Total Lactic Acid (ug/mL)', hue='Treatment', kind='bar', aspect=2.2)
plot.set(ylim=(0, 6000), ylabel='Lactic Acid (μg/mL)', xlabel='')
plot.set_xticklabels(rotation=30, ha='right')
#plot.savefig(figPath.joinpath('lactic_acid_production_v2.png'), dpi=300, bbox_inches='tight')
```

```python
plot = sns.catplot(data=lactateData, x='Species', y='Ratio', hue='Treatment', kind='bar', aspect=2.2)
plot.set(ylabel='Ratio (L-Lactate:D-Lactate)', xlabel='')
plot.set_xticklabels(rotation=30, ha='right')
#plot.savefig(figPath.joinpath('lactate_ratios.png'), dpi=300, bbox_inches='tight')
```
